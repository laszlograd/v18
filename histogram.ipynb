```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def strip(text):
    try:
        return text.strip()
    except AttributeError:
        return text

def make_int(text):
    return int(text)

data = pd.read_csv('resz.csv', converters = {'AE' : make_int, 'FE' : make_int})
data['perc'] = round(data['FE'] / data['AE'] * 100)
```

```python
data
```

```python
%matplotlib inline
data.hist(column="perc", bins=120, figsize=(18, 16))
```

```python
from scipy.stats import kurtosis
kurtosis(data['perc'])
```

```python
from scipy.stats import skew
skew(data['perc'])
```
